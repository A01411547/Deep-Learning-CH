```python
model = load_model('C:/Users/Hunter/Downloads/perritos_y_ygatitos.h5')
model
```

```python
#
from tensorflow.keras.utils import load_img
from tensorflow.keras.utils import img_to_array
import numpy as np
input= load_img('C:/Users/Hunter/Downloads/pruebas/18.jpg', target_size=(224, 224))
im= np.array(input)
im= im/ 224.0
im= im.reshape(1,224,224,3)
predict= model.predict(im)
print(predict[0])
```
